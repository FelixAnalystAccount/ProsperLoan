## Univariate Exploration

I will start by looking at the distribution of the main point of interest: OnTimeProsperPayments

```python
# start with a standard-scaled plot of OnTimeProsperPayment
binsize = 4
bins = np.arange(0,prosper_loan['OnTimeProsperPayments'].max()+binsize, binsize)
plt.figure(figsize=[8,5])
plt.hist(data = prosper_loan, x = 'OnTimeProsperPayments', bins = bins)
plt.xlabel('OnTimeProsperPayments')
plt.ylabel('Number of OnTimePayment made')
plt.title('OnTimeProsperPayments on Standard scale');
```

The distribution is right skewed with long tail. I will try to plot it on a log scale.

```python
# The plot is skewed to the right so i'll plot it on a log scale
log_binsize = 0.05
bins = 10 ** np.arange(0, np.log10(prosper_loan['OnTimeProsperPayments'].max())+log_binsize, log_binsize)
plt.figure(figsize =[8,5])
plt.hist(data = prosper_loan, x = 'OnTimeProsperPayments', bins = bins)
plt.title('OnTimeProsperPayments on log scale')
plt.xscale('symlog')
# plt.yscale('log')
plt.xlim([0,50])
plt.xticks([0, 0.01e2,0.05e2, 0.1e2, 0.5e2, 0.1e3, 0.5e3], [0, '1','5', '10', '50', '100', '500'])
plt.xlabel('OnTimeProsperPayments')
plt.ylabel('Number of OnTimePayment made');
```

To start of with, I want to look at the pairwise correlations between features in the data.

```python
numeric_vars =  ['IncomeRange','EmploymentStatusDuration','LoanOriginalAmount','StatedMonthlyIncome','MonthlyLoanPayment','DebtToIncomeRatio','BorrowerAPR','OnTimeProsperPayments', 'Term']
categoric_vars = ['ProsperRating (Alpha)']
```

```python
# correlation plot
plt.figure(figsize = [8,5])
sb.heatmap(data = prosper_loan[numeric_vars].corr(), annot = True, fmt = '.3f',cmap = 'vlag_r', center = 0)
plt.title('Correlation Matrix')
plt.show()
```

```python
samples = np.random.choice(prosper_loan.shape[0], 500, replace = False)
prosper_loan_samp = prosper_loan.loc[samples,:]

g = sb.PairGrid(data = prosper_loan, vars = numeric_vars)
g = g.map_diag(plt.hist, bins = 20);
g.map_offdiag(plt.scatter);
```

There is a positive correlation between MonthlyLoanPayment and LoanOriginalAmount and a slight positive correlation between Term and LoanOriginalAmount. There are negative correlation between Term and MonthlyLoanPayment, BorrowerAPR and Term and DebtToIncomeRatio and Term.
