```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
prosper_loan = pd.read_csv('prosperLoanData.csv')
```

## Distribution of OnTimeProsperPayments on log scale 

A look into The distribution of the OnTimeProsperPayment showed a right skewed distribution with long tail on a standard scale. It was plotted on the log scale for better visualization.

```python
# The plot is skewed to the right so i'll plot it on a log scale
log_binsize = 0.05
bins = 10 ** np.arange(0, np.log10(prosper_loan['OnTimeProsperPayments'].max())+log_binsize, log_binsize)
plt.figure(figsize =[8,5])
plt.hist(data = prosper_loan, x = 'OnTimeProsperPayments', bins = bins)
plt.title('OnTimeProsperPayments on log scale')
plt.xscale('symlog')
# plt.yscale('log')
plt.xlim([0,50])
plt.xticks([0, 0.01e2,0.05e2, 0.1e2, 0.5e2, 0.1e3, 0.5e3], [0, '1','5', '10', '50', '100', '500'])
plt.xlabel('OnTimeProsperPayments')
plt.ylabel('Number of OnTimePayment made');
```

## LoanOriginalAmount Vs MonthlyLoanPayment

The distribution showing a positive correlation between these two features. It showed the highest LoanOriginalAmount of $35000 has a MonthlyLoanPayment between $0 - 1000

```python
#  A plot of LoanOriginalAmount and MonthlyLoanPayment
plt.figure(figsize = [8,8])
plt.scatter(data = prosper_loan, x = 'MonthlyLoanPayment', y = 'LoanOriginalAmount', alpha = 1/10)
plt.xlabel('MonthlyLoanPayment')
plt.ylabel('LoanOriginalAmount')
plt.title('LoanOriginalAmount and MonthlyLoanPayment Distribution')
plt.show()
```

## MonthlyLoanPayment vs OnTimeProsperPayments vs ProsperScore

The distribution showed a positive corellation between OnTimeProsperPayment and MonthlyLoanPayment as the ProsperScore increases.

```python
plt.scatter(data = prosper_loan, x = 'LoanOriginalAmount', y = 'MonthlyLoanPayment', c = 'ProsperScore', cmap = 'inferno_r')
plt.colorbar(label = 'ProsperScore')
plt.xlabel('OnTimeProsperPayments')
plt.ylabel('MonthlyLoanPayment')
plt.title('MonthlyLoanPayment,OnTimeProsperPayments and ProsperScore');
```
